```python
# Lists - an ordered collection of items
songs = ["Bohemian Rhapsody", "Hotel California", "Smells Like Teen Spirit", "Billie Jean", "Lose Yourself"]

print(songs)
print(type(songs))
print(len(songs))
```

```python
# Dictionaries - like a labeled box for each song
# Instead of just a name, we store multiple details together

song = {
    "title": "Bohemian Rhapsody",
    "artist": "Queen",
    "year": 1975,
    "plays": 1200
}

print(song)
print(song["title"])
print(song["plays"])
```

```python
# A list of dictionaries - this is basically a table
# Each dictionary is one row, each key is a column name

music_data = [
    {"title": "Bohemian Rhapsody", "artist": "Queen", "year": 1975, "plays": 1200},
    {"title": "Hotel California", "artist": "Eagles", "year": 1977, "plays": 950},
    {"title": "Smells Like Teen Spirit", "artist": "Nirvana", "year": 1991, "plays": 1100},
    {"title": "Billie Jean", "artist": "Michael Jackson", "year": 1982, "plays": 1500},
    {"title": "Lose Yourself", "artist": "Eminem", "year": 2002, "plays": 870}
]

# Get the first song
print(music_data[0])

# Get just the title of the third song
print(music_data[2]["title"])

# How many songs total
print(len(music_data))
```

```python
# A loop goes through every item in a list one by one
# 'song' is a temporary variable name - it holds one dictionary at a time

for song in music_data:
    print(song["title"], "-", song["artist"])
```

```python
# We can also do calculations inside a loop
for song in music_data:
    if song["plays"] > 1000:
        print(song["title"], "is popular with", song["plays"], "plays")
```

```python
# A function is a reusable block of code you give a name to
# def means "define" - you are defining what the function does
# The function does nothing until you actually CALL it

def describe_song(song):
    print(song["title"], "by", song["artist"])
    print("Released in", song["year"])
    print("Total plays:", song["plays"])
    print("---")

# Calling the function - this is where it actually runs
describe_song(music_data[0])
describe_song(music_data[3])

# "A function takes some input, does something with it, and optionally gives something back. 
# It means you write the logic once and reuse it instead of copy pasting."
```

```python
# print() just displays something on screen
# return actually gives a value BACK so you can use it later

def get_popularity(song):
    if song["plays"] > 1000:
        return "popular"
    else:
        return "average"

# Now we can USE the result, not just print it
result = get_popularity(music_data[0])
print(result)

# Or use it directly inside a loop
for song in music_data:
    label = get_popularity(song)
    print(song["title"], "-", label)
```

```python
import csv

# Writing our music data to a CSV file
with open("music_data.csv", "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=["title", "artist", "year", "plays"])
    writer.writeheader()
    writer.writerows(music_data)

print("File saved!")

# Reading it back
with open("music_data.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row["title"], "-", row["plays"], "plays")
```
