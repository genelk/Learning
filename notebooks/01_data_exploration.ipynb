```python
# Exploring columns in the data files

files = [
    "data/data.csv",
    "data/data_by_genres.csv",
    "data/data_w_genres.csv",
    "data/data_by_artist.csv", 
    "data/data_by_year.csv",
]

for f in files:
    df = load_from_zip("data/archive.zip", f)
    print(f"\n{f}")
    print(f"Shape: {df.shape}")
    print(f"Columns: {df.columns.tolist()}")
```

```python
unique_genres = df['genres'].unique()

print("First 20 unique genres:")
unique_genres

total_unique = len(unique_genres)
print(f"There are {total_unique} unique genres.")

# Count actual missing (Null) values
null_count = df['genres'].isnull().sum()

# Find rows that are just empty brackets '[]'
empty_bracket_count = len(df[df['genres'] == '[]'])

print(f"Actual Nulls: {null_count}")
print(f"Empty Brackets '[]': {empty_bracket_count}")

# Check if any ID or Song Name + Artists combination is repeated
# duplicate_count = df.duplicated(subset=['id']).sum()
# print(f"Number of duplicate IDs: {duplicate_count}")
```

```python
with pd.option_context('display.max_columns', None, 'display.max_rows', None):
    print(df.describe())
```

```python
# Validating Audio Ranges
quiet_tracks = df[df['loudness'] == -60.0]
print(f"Found {len(quiet_tracks)} super quiet tracks.")
print(quiet_tracks)
```

```python
# Look at tracks where speechiness is high
audiobook_check = df[df['duration_ms'] > 0.9]
#audiobook_check.head(20)

# Find tracks longer than 15 mins (900,000 ms) AND popularity less than 5
hidden_books = df[ (df['duration_ms'] > 900000) & (df['speechiness'] > 0.8) ]
hidden_books.head(20)
```
